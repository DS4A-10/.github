## Creating Country Poverty Dataframe - PR dataframe

```python
PR =pd.read_csv("Country Poverty Rates.csv",header = 4)
PR.shape
```

```python
# pd.set_option('display.max_columns', None)
# PR.head(100)
```

```python
# pd.set_option("max_rows", None)
# PR
```

```python
PR['Indicator Name'].unique()
```

```python
# Selecting the indicator tu use for analysis: 
# 'Poverty headcount ratio at national poverty lines (% of population)' - using Indicator https://data.worldbank.org/indicator/SI.POV.NAHC?locations=CO&view=map&year_high_desc=true
PR = PR[(PR['Indicator Name'] == 'Poverty headcount ratio at national poverty lines (% of population)')][['Country Name','2017','2018','2019','2020']]
```

```python
print(PR.shape)
PR.head(3)
```

```python
# Setting index to country name to facilitate selection of most recent year value from dataset
PR = PR.set_index('Country Name')
```

```python
PR.head(3)
```

```python
# using last non nan value from the years between (2017-2020)
PR['Most recent year value (2017-2020)_Poverty headcount ratio at national poverty lines (% of population)'] = PR.ffill(axis=1).iloc[:, -1]
PR.head(3)
```

```python
# Resetting index
PR = PR.reset_index()
```

```python
PR.head(2)
```

```python
# Selecting only columns to be used in final dataframe (to merge with regions data)
PR = PR.filter(['Country Name', 'Most recent year value (2017-2020)_Poverty headcount ratio at national poverty lines (% of population)'], axis=1)
PR.head()
```
